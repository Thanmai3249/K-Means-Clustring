# K-Means Algorithm implementation
Unsupervised learning algorithm K-Means Clustering divides the unlabeled dataset into various clusters. In this case, K specifies how many pre-defined clusters or groups must be produced during the process; for example, if K=5, there will be 5 clusters, if K=10, there will be 10 clusters, and so on. The two major functions of the k-means clustering algorithm are:

- Chooses the K centre points' ideal value.
- Each data point is matched with the nearest k-center. By calculating the separation between the k centre points and the data points, groups assign based on the k centre points.

# Using the elbow method to find the num of clusters

WCSS (Within-Cluster Sum of Squares) is an evaluation metric used in K-Means clustering. It measures the sum of squared distances between all the data points in a cluster and its centroid. By minimizing the WCSS, K-Means is able to identify clusters with the most homogeneous data points.

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i, init='k-means++',max_iter=300,n_init=10,random_state=0)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("no. of Cluster")
plt.ylabel("wcss")
plt.show()
```

WCSS doesn’t reduce much after k=5. So, we can choose 5 as the perfect K value or Clusters.
